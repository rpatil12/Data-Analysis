```python
# Structure of your solution to Assignment 1 

import string

def tokenize(text):
    
    # split by space (including \tab and \n)
    tokens=text.split()
    
    # initialize a list
    cleaned_tokens = []
    
    
    for token in tokens:
        
        # remove space around each token
        token = token.strip("")
        
        # recursively remove punctuations around a token
        while len(token)>0:  
            
            # remove leading punctuation
            if token[0] in string.punctuation:
                token = token[1:]
                
            # remove trailing punctuation
            if token[-1] in string.punctuation:
                token = token[0:len(token)-1]
            
            # break if no leading or trailing punctuation
            if not (token[0] in string.punctuation or token[-1] in string.punctuation):
                break
        
        # add non-empty token in lower case
        if len(token)>0:
            cleaned_tokens.append(token.lower())
    
    return cleaned_tokens


class Text_Analyzer(object):
    
    def __init__(self, doc):
        
        # initialize two attributes
        self.input_string=doc
        self.token_count ={}
        
    def analyze(self):
        
        # tokenize input string
        tokens= tokenize(self.input_string)
        
        # get token count
        self.token_count = {w:tokens.count(w) for w in set(tokens)}
        
            
    def topN(self, N):
        
        result =[]
        
        if N>0:
            # sort count dictionary by value in descending order
            result = sorted(self.token_count.items(), \
                            key=lambda item:-item[1])[0:N]
        
        return result

def bigram(doc, N):
    
    result=[]
   
    # token input document
    tokens=tokenize(doc)
    
    # initialize bigram count dictionary
    bigrams_count = {}
    
    if len(tokens)>1:
        # loop through to the last second token
        for i in range(len(tokens)-1):
            # create count dictionary for bigrams
            if (tokens[i], tokens[i+1]) in bigrams_count:
                bigrams_count[(tokens[i], tokens[i+1])]+=1
            else:
                bigrams_count[(tokens[i], tokens[i+1])]=1
    
    if N>1 and len(bigrams_count)>0:
        # sort dictionary by value in descending order
        result = sorted(bigrams_count.items(), \
                         key=lambda item:-item[1])[0:N]
    
    return result

# best practice to test your class
# if your script is exported as a module,
# the following part is ignored
# this is equivalent to main() in Java

if __name__ == "__main__":  
    
    # Test Question 1
    text=''' There was nothing so VERY remarkable in that; nor did Alice
think it so VERY much out of the way to hear the Rabbit say to
itself, `Oh dear!  Oh dear!  I shall be late!'  (when she thought
it over afterwards, it occurred to her that she ought to have
wondered at this, but at the time it all seemed quite natural);
but when the Rabbit actually TOOK A WATCH OUT OF ITS WAISTCOAT-
POCKET, and looked at it, and then hurried on, Alice started to
her feet, for it flashed across her mind that she had never
before seen a rabbit with either a waistcoat-pocket, or a watch to
take out of it, and burning with curiosity, she ran across the
field after it, and fortunately was just in time to see it pop
down a large rabbit-hole under the hedge.
''' 
    print("Test Question 1")
    print(tokenize(text))
    
    
    # Test Question 2
    print("\nTest Question 2")
    analyzer=Text_Analyzer(text)
    analyzer.analyze()
    print(analyzer.token_count)
    print(analyzer.topN(10))
    
    # Note that top words are not very meaningful. 
    # They are called stop words
    
    #3 Test Question 3
    print("\nTest Question 3")
    print(bigram(text,5))
    
    # Note that top bigrams may not necessarily \
    # be meaningful phrases
    
```
